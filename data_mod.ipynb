```python
import pandas as pd
pd.options.mode.chained_assignment = None 
# https://stackoverflow.com/questions/20625582/how-to-deal-with-settingwithcopywarning-in-pandas
```

```python
covid = pd.read_csv("data\covid_deaths_monthly.csv")
covid
```

```python
covid_mod = covid[covid['Year Code'] > 2004]
covid_mod.drop(covid_mod.columns[[2,4,5,12]], axis=1, inplace=True)
covid_mod.rename(columns={'Year Code':'Month','Crude Rate':'Infection'}, inplace=True)
covid_mod['Month'].replace(2005, 'January', inplace=True)
covid_mod['Month'].replace(2006, 'February', inplace=True)
covid_mod['Month'].replace(2007, 'March', inplace=True)
covid_mod['Month'].replace(2008, 'April', inplace=True)
covid_mod['Month'].replace(2009, 'May', inplace=True)
covid_mod['Month'].replace(2010, 'June', inplace=True)
covid_mod['Month'].replace(2011, 'July', inplace=True)
covid_mod['Month'].replace(2012, 'August', inplace=True)
covid_mod['Month'].replace(2013, 'September', inplace=True)
covid_mod['Month'].replace(2014, 'October', inplace=True)
covid_mod['Month'].replace(2015, 'November', inplace=True)
covid_mod['Month'].replace(2016, 'December', inplace=True)
covid_mod
# covid['Year Code'].max()
# pd.unique(covid['Year Code'])
```
